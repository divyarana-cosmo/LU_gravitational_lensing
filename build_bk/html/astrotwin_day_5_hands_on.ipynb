# Day 5 - Environmental dependence of matter profiles

In this session, we will study the weak lensing signal around the satellite galaxies in a cluster environment. We will first make some scatter plots and decide on the selection criteria for our satellite galaxies. As before, we will then use them as lenses and measure the weak-lensing signal around them. 

The catalog data is available by the name "redmapper_satellites.dat"

## Lets first make a scatter plot

Here we will make the plot between redshift and projected separation of satellite from the cluster center. We will then select the sample using the cuts on the separation.
- **lenses: /home/idies/workspace/Storage/divyar/AstroTwin_Colo_2024/DataStore/redmapper_satellites.dat**

```python
#loading the required packages
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# we are setting the below params to adjust plots 
plt.rcParams['figure.figsize'] = [3,3]
plt.rcParams['figure.dpi'] = 250
```

```python
#putting the path of the lens catalog
len_path = '/home/idies/workspace/Storage/divyar/AstroTwin_Colo_2024/DataStore/redmapper_satellites.dat'

lenses = pd.read_csv(len_path, delim_whitespace=1)

#printing the columns in the file
print(lenses.keys())
plt.plot(lenses['zred'], lenses['rsep[h-1Mpc]'], '.', ms=0.1)
plt.plot(np.linspace(0.1,0.33,10),0.1*np.ones(10) , 'r-')
plt.plot(np.linspace(0.1,0.33,10),0.4*np.ones(10) , 'r-')
plt.plot(0.1*np.ones(10), np.linspace(0.1,0.4,10) ,  'r-')
plt.plot(0.33*np.ones(10), np.linspace(0.1,0.4,10) , 'r-')

plt.xlabel(r'$z_{\rm red}$')
plt.ylabel(r'$R_{\rm sep} [h^{-1} Mpc]$')

#printing the number of lenses in a selection cut
idx = (lenses['zred']>0.1) & (lenses['zred']<0.33) & (lenses['rsep[h-1Mpc]']>0.1) & (lenses['rsep[h-1Mpc]']<0.4)
print('no of bins in first selection cut : %d'%np.sum(idx))
```
